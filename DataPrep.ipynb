```
import torch
import torch.nn as nn
import os

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torch.utils.data import Dataset, DataLoader
import imageio
```

```
trainData = os.listdir(trainPath)
print(len(trainData))
print(trainData[0])
```

```
testData = os.listdir(testPath)
print(len(testData))
```

```
traindepthData = os.listdir(traindepthPath)
print(len(traindepthData))
```

```
#Reading Training Data
n =0
training_size = 9800*4
train_images = np.zeros((training_size, 104, 104))
for i in sorted(trainData):
  for j in range(1,5):
    img_arr = imageio.imread(trainPath+i+'/'+str(j)+'.png')
    train_images[n] = img_arr
    n += 1
train_images /= 255.0
```

```
#Reading Testing Data
n =0
testing_size = 150*4
test_images = np.zeros((testing_size, 104, 104))
for i in sorted(testData):
  for j in range(1,5):
    img_arr = imageio.imread(testPath+i+'/'+str(j)+'.png')
    test_images[n] = img_arr
    n += 1
test_images /= 255.0
```

```
#Reading Traindepth Data
n =0
traindepth_size = 9800
traindepth_images = np.zeros((traindepth_size, 104, 104))
for i in sorted(traindepthData):
  img_arr = imageio.imread(traindepthPath+i+'/depth.png')
  traindepth_images[n] = img_arr
  n += 1
traindepth_images /= 255.0
```

```
#save to numpy file
np.save('trainImages.npy', train_images)
np.save('testImages.npy', test_images)
np.save('traindepthImages.npy', traindepth_images)
```
